```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = 224
BATCH_SIZE = 32
DATASET_PATH = r"D:\Lohith\BTECH CODES\Machine_Learning\Project\Crop_Disease\plantvillage-dataset\plantvillage dataset\color"
datagen = ImageDataGenerator(rescale=1./255)

data = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training"
)

print("Number of classes:", data.num_classes)
print("Class labels (first 10):", list(data.class_indices.keys())[:10])
```

```python
import os
import shutil

SOURCE_DIR = r"D:\Lohith\BTECH CODES\Machine_Learning\Project\Crop_Disease\plantvillage-dataset\plantvillage dataset\color"
TARGET_DIR = r"D:\Lohith\BTECH CODES\Machine_Learning\Project\Crop_Disease\dataset_filtered"

os.makedirs(TARGET_DIR, exist_ok=True)

for folder in os.listdir(SOURCE_DIR):
    if folder.startswith("Tomato") or folder.startswith("Potato"):
        src = os.path.join(SOURCE_DIR, folder)
        dst = os.path.join(TARGET_DIR, folder)
        if not os.path.exists(dst):
            shutil.copytree(src, dst)

print("✅ Tomato and Potato folders copied successfully!")
```
